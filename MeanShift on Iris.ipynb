```python
from sklearn import datasets
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import MeanShift

import matplotlib.pyplot as plt
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D


import pandas as pd
```

```python
#loading data
iris = datasets.load_iris()
df = pd.DataFrame(iris.data,columns=["sepal_length","sepal_width","petal_length","petal_width"])
```

```python

df.head()
```

```python
#standardizing the features
scalar = StandardScaler()
X_std = scalar.fit_transform(df)
```

```python
X_std
```

```python
mObj = MeanShift(n_jobs = -1)
model = mObj.fit(X_std)
```

```python
cluster_center = mObj.cluster_centers_
```

```python
mObj.labels_
```

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d",)
ax.scatter(X_std[:,0],X_std[:,1],marker='o')
ax.scatter(cluster_center[:,0],cluster_center[:,1],marker = 'x',color = 'red',linewidth = 5,zorder = 10)
plt.show()
```

In above figure there are two clusters , sepal_length,sepal_width and Mshift algorithm fins two centroids of these two clusters(red color)

```python
fig = plt.figure()
ax = fig.add_subplot(111,projection = "3d")
ax.scatter(X_std[:,2],X_std[:,3],marker='o',color = "grey")
ax.scatter(cluster_center[:,2],cluster_center[:,3],marker = 'x',color = 'red',linewidth = 5,zorder = 10)
plt.show()
```

In above figure there are two clusters , petal_length,petal_width and Mshift algorithm fins two centroids of these two clusters(red color).Please look carefully centers are visible
